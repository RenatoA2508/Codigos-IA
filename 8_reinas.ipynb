```python
from ortools.sat.python import cp_model
```

```python
class ImprimirSoluciones(cp_model.CpSolverSolutionCallback):
    """
    La clase es necesaria para poder obtener todas las soluciones posibles del modelo
    """
    def __init__(self, q):
        cp_model.CpSolverSolutionCallback.__init__(self)
        self.q = q
        self.contador = 0

    def OnSolutionCallback(self):
        self.contador += 1
        solucion = [self.Value(self.q[i]) for i in range(8)]
        print(f"SoluciĂ³n #{self.contador}: {solucion}")
        for fila in range(8):
            for col in range(8):
                if solucion[col] == fila:
                    print("Q", end=" ")
                else:
                    print(".", end=" ")
            print()
        print()

modelo = cp_model.CpModel()

q = [modelo.NewIntVar(0, 7, f'q{i}') for i in range(8)]

modelo.AddAllDifferent(q)

for i in range(8):
    for j in range(i + 1, 8):
        modelo.Add(q[i] - q[j] != i - j)
        modelo.Add(q[i] - q[j] != j - i)

solver = cp_model.CpSolver()
impresor = ImprimirSoluciones(q)

solver.SearchForAllSolutions(modelo, impresor)

```
