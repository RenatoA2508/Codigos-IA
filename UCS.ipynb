```python
#Participación
import heapq
grafo = {
    'Arad': {
        'Zerind': 75,
        'Sibiu': 140,
        'Timisoara': 118
    },
    'Zerind': {
        'Arad': 75,
        'Oradea': 71
    },
    'Oradea': {
        'Zerind': 71,
        'Sibiu': 151
    },
    'Sibiu': {
        'Arad': 140,
        'Oradea': 151,
        'Fagaras': 99,
        'Rimnicu Vilcea': 80
    },
    'Timisoara': {
        'Arad': 118,
        'Lugoj': 111
    },
    'Lugoj': {
        'Timisoara': 111,
        'Mehadia': 70
    },
    'Mehadia': {
        'Lugoj': 70,
        'Drobeta': 75
    },
    'Drobeta': {
        'Mehadia': 75,
        'Craiova': 120
    },
    'Craiova': {
        'Drobeta': 120,
        'Rimnicu Vilcea': 146,
        'Pitesti': 138
    },
    'Rimnicu Vilcea': {
        'Sibiu': 80,
        'Craiova': 146,
        'Pitesti': 97
    },
    'Fagaras': {
        'Sibiu': 99,
        'Bucharest': 211
    },
    'Pitesti': {
        'Rimnicu Vilcea': 97,
        'Craiova': 138,
        'Bucharest': 101
    },
    'Bucharest': {
        'Fagaras': 211,
        'Pitesti': 101,
        'Giurgiu': 90,
        'Urziceni': 85
    },
    'Giurgiu': {
        'Bucharest': 90
    },
    'Urziceni': {
        'Bucharest': 85,
        'Hirsova': 98,
        'Vaslui': 142
    },
    'Hirsova': {
        'Urziceni': 98,
        'Eforie': 86
    },
    'Eforie': {
        'Hirsova': 86
    },
    'Vaslui': {
        'Urziceni': 142,
        'Iasi': 92
    },
    'Iasi': {
        'Vaslui': 92,
        'Neamt': 87
    },
    'Neamt': {
        'Iasi': 87
    }
}

def ucs(grafo, inicio, fin):
    cola = [(0, inicio)]  # (costo, nodo)
    visitados = set()
    ruta = {inicio: None}
    
    while cola:
        print(f"Cola: {cola}")  # Imprimir el contenido de la cola
        costo, nodo = heapq.heappop(cola)
        
        if nodo not in visitados:
            visitados.add(nodo)
            
            if nodo == fin:
                return costo, ruta
            
            for vecino, costo_arista in grafo[nodo].items():
                nuevo_costo = costo + costo_arista
                
                if vecino not in visitados:
                    ruta[vecino] = nodo
                    heapq.heappush(cola, (nuevo_costo, vecino))
    
    return float("inf"), {}


# Llamamos a la función
costo, ruta = ucs(grafo, 'Arad', 'Bucharest')

# Imprimimos la ruta más corta
camino = []
nodo = 'Bucharest'

while nodo is not None:
    camino.append(nodo)
    nodo = ruta[nodo]

camino.reverse()

print("La ruta más corta es:", camino)
print("El costo es:", costo)
```

```python
import heapq
import networkx as nx
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D

# =========================
# GRAFO DE RUMANIA
# =========================
grafo = {
    'Arad': {
        'Zerind': 75,
        'Sibiu': 140,
        'Timisoara': 118
    },
    'Zerind': {
        'Arad': 75,
        'Oradea': 71
    },
    'Oradea': {
        'Zerind': 71,
        'Sibiu': 151
    },
    'Sibiu': {
        'Arad': 140,
        'Oradea': 151,
        'Fagaras': 99,
        'Rimnicu Vilcea': 80
    },
    'Timisoara': {
        'Arad': 118,
        'Lugoj': 111
    },
    'Lugoj': {
        'Timisoara': 111,
        'Mehadia': 70
    },
    'Mehadia': {
        'Lugoj': 70,
        'Drobeta': 75
    },
    'Drobeta': {
        'Mehadia': 75,
        'Craiova': 120
    },
    'Craiova': {
        'Drobeta': 120,
        'Rimnicu Vilcea': 146,
        'Pitesti': 138
    },
    'Rimnicu Vilcea': {
        'Sibiu': 80,
        'Craiova': 146,
        'Pitesti': 97
    },
    'Fagaras': {
        'Sibiu': 99,
        'Bucharest': 211
    },
    'Pitesti': {
        'Rimnicu Vilcea': 97,
        'Craiova': 138,
        'Bucharest': 101
    },
    'Bucharest': {
        'Fagaras': 211,
        'Pitesti': 101,
        'Giurgiu': 90,
        'Urziceni': 85
    },
    'Giurgiu': {
        'Bucharest': 90
    },
    'Urziceni': {
        'Bucharest': 85,
        'Hirsova': 98,
        'Vaslui': 142
    },
    'Hirsova': {
        'Urziceni': 98,
        'Eforie': 86
    },
    'Eforie': {
        'Hirsova': 86
    },
    'Vaslui': {
        'Urziceni': 142,
        'Iasi': 92
    },
    'Iasi': {
        'Vaslui': 92,
        'Neamt': 87
    },
    'Neamt': {
        'Iasi': 87
    }
}

# =========================
# UCS CON REGISTRO DE PROCESO
# =========================
def ucs_visual(grafo, inicio, fin):
    cola = [(0, inicio)]  # (costo_acumulado, nodo)
    visitados = set()
    padres = {inicio: None}
    mejor_costo = {inicio: 0}

    aristas_exploradas = set()   # aristas aceptadas hacia nuevos mejores caminos
    aristas_descartadas = set()  # aristas descartadas
    historial_descartes = []     # para imprimir explicación

    paso = 1

    while cola:
        print("\n" + "=" * 50)
        print(f"PASO {paso}")
        print("Cola antes de extraer:", sorted(cola))

        costo, nodo = heapq.heappop(cola)
        print(f"Se extrae: ({costo}, '{nodo}')")

        # Si es una entrada vieja de la cola, se ignora
        if costo > mejor_costo.get(nodo, float("inf")):
            print(f"-> Se ignora porque ya existe un mejor costo para {nodo}.")
            paso += 1
            continue

        if nodo in visitados:
            print(f"-> {nodo} ya fue visitado.")
            paso += 1
            continue

        visitados.add(nodo)
        print("Visitados:", visitados)

        if nodo == fin:
            print(f"\nSe llegó al destino: {fin}")
            break

        for vecino, costo_arista in grafo[nodo].items():
            nuevo_costo = costo + costo_arista

            # Caso 1: vecino ya visitado -> descartar
            if vecino in visitados:
                arista = tuple(sorted((nodo, vecino)))
                aristas_descartadas.add(arista)
                historial_descartes.append(
                    f"Descartado {nodo} -> {vecino} porque {vecino} ya fue visitado."
                )
                print(f"Descartado {nodo} -> {vecino} (ya visitado)")
                continue

            # Caso 2: se encontró mejor camino
            if nuevo_costo < mejor_costo.get(vecino, float("inf")):
                mejor_costo[vecino] = nuevo_costo
                padres[vecino] = nodo
                heapq.heappush(cola, (nuevo_costo, vecino))

                arista = tuple(sorted((nodo, vecino)))
                aristas_exploradas.add(arista)

                print(f"Se agrega a la cola: ({nuevo_costo}, '{vecino}')")
            else:
                # Caso 3: el camino no mejora -> descartar
                arista = tuple(sorted((nodo, vecino)))
                aristas_descartadas.add(arista)
                historial_descartes.append(
                    f"Descartado {nodo} -> {vecino} porque el costo {nuevo_costo} no mejora "
                    f"el mejor costo conocido ({mejor_costo.get(vecino)})."
                )
                print(f"Descartado {nodo} -> {vecino} (no mejora costo)")

        print("Cola después de expandir:", sorted(cola))
        paso += 1

    # Reconstruir camino
    if fin not in padres:
        return float("inf"), [], visitados, aristas_exploradas, aristas_descartadas, historial_descartes

    camino = []
    nodo = fin
    while nodo is not None:
        camino.append(nodo)
        nodo = padres[nodo]
    camino.reverse()

    return mejor_costo[fin], camino, visitados, aristas_exploradas, aristas_descartadas, historial_descartes


# =========================
# DIBUJAR EL GRAFO
# =========================
def dibujar_grafo_proceso(grafo, camino, visitados, aristas_exploradas, aristas_descartadas, inicio, fin):
    G = nx.Graph()

    for ciudad, vecinos in grafo.items():
        for vecino, peso in vecinos.items():
            G.add_edge(ciudad, vecino, weight=peso)

    # Posiciones manuales para que se parezca al mapa de Rumania
    pos = {
        'Arad': (1, 5),
        'Zerind': (2, 6),
        'Oradea': (3, 7),
        'Sibiu': (5, 5),
        'Timisoara': (1, 3),
        'Lugoj': (3, 2),
        'Mehadia': (3, 1),
        'Drobeta': (3, 0),
        'Craiova': (6, 0),
        'Rimnicu Vilcea': (5, 3),
        'Fagaras': (8, 4.8),
        'Pitesti': (8, 2.2),
        'Bucharest': (11, 1.5),
        'Giurgiu': (10.5, -0.5),
        'Urziceni': (13, 1.8),
        'Hirsova': (15, 1.8),
        'Eforie': (16, 0),
        'Vaslui': (15, 4),
        'Iasi': (14, 6),
        'Neamt': (12, 7)
    }

    # Aristas de la ruta final
    aristas_camino = set()
    for i in range(len(camino) - 1):
        aristas_camino.add(tuple(sorted((camino[i], camino[i + 1]))))

    # Dibujar todas las aristas base
    plt.figure(figsize=(16, 9))
    nx.draw_networkx_edges(G, pos, edge_color='lightgray', width=1.5)

    # Dibujar aristas exploradas (excepto las de la ruta final)
    exploradas_sin_ruta = list(aristas_exploradas - aristas_camino)
    nx.draw_networkx_edges(
        G, pos,
        edgelist=exploradas_sin_ruta,
        edge_color='orange',
        width=3
    )

    # Dibujar aristas descartadas (excepto las de la ruta final)
    descartadas_sin_ruta = list(aristas_descartadas - aristas_camino)
    nx.draw_networkx_edges(
        G, pos,
        edgelist=descartadas_sin_ruta,
        edge_color='red',
        style='dashed',
        width=2.5
    )

    # Dibujar ruta más corta
    nx.draw_networkx_edges(
        G, pos,
        edgelist=list(aristas_camino),
        edge_color='green',
        width=4
    )

    # Colorear nodos
    colores_nodos = []
    for nodo in G.nodes():
        if nodo == inicio:
            colores_nodos.append('skyblue')
        elif nodo == fin:
            colores_nodos.append('lightcoral')
        elif nodo in camino:
            colores_nodos.append('lightgreen')
        elif nodo in visitados:
            colores_nodos.append('khaki')
        else:
            colores_nodos.append('white')

    nx.draw_networkx_nodes(
        G, pos,
        node_color=colores_nodos,
        edgecolors='black',
        node_size=900
    )

    nx.draw_networkx_labels(G, pos, font_size=10, font_weight='bold')

    # Etiquetas de pesos
    edge_labels = nx.get_edge_attributes(G, 'weight')
    nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=9)

    # Leyenda
    leyenda = [
        Line2D([0], [0], color='green', lw=4, label='Ruta más corta'),
        Line2D([0], [0], color='orange', lw=3, label='Caminos explorados'),
        Line2D([0], [0], color='red', lw=2.5, linestyle='--', label='Caminos descartados'),
        Line2D([0], [0], marker='o', color='w', markerfacecolor='skyblue', markeredgecolor='black', markersize=12, label='Inicio'),
        Line2D([0], [0], marker='o', color='w', markerfacecolor='lightcoral', markeredgecolor='black', markersize=12, label='Destino'),
        Line2D([0], [0], marker='o', color='w', markerfacecolor='khaki', markeredgecolor='black', markersize=12, label='Visitados'),
    ]

    plt.legend(handles=leyenda, loc='upper left', fontsize=10)
    plt.title("Visualización de UCS: caminos explorados, descartados y ruta más corta", fontsize=14)
    plt.axis('off')
    plt.tight_layout()
    plt.show()


# =========================
# EJECUCIÓN
# =========================
inicio = 'Arad'
fin = 'Bucharest'

costo, camino, visitados, aristas_exploradas, aristas_descartadas, historial_descartes = ucs_visual(grafo, inicio, fin)

print("\n" + "=" * 50)
print("RESULTADO FINAL")
print("La ruta más corta es:", camino)
print("El costo es:", costo)

print("\nCAMINOS DESCARTADOS:")
for d in historial_descartes:
    print("-", d)

dibujar_grafo_proceso(
    grafo,
    camino,
    visitados,
    aristas_exploradas,
    aristas_descartadas,
    inicio,
    fin
)
```
